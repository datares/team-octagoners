```python
from sklearn import preprocessing
def prep_ufc(data_df):
    df = data_df.copy()
    df = df[['no_of_rounds', 'B_losses', 'B_total_rounds_fought',
       'B_total_title_bouts', 'B_win_by_Decision_Majority',
       'B_win_by_Decision_Split', 'B_win_by_Decision_Unanimous',
       'B_win_by_KO/TKO', 'B_win_by_Submission',
       'B_win_by_TKO_Doctor_Stoppage', 'B_wins', 'B_Height_cms',
       'B_Reach_cms', 'B_Weight_lbs',
       'R_losses', 'R_total_rounds_fought', 'R_total_title_bouts',
       'R_win_by_Decision_Majority', 'R_win_by_Decision_Split',
       'R_win_by_Decision_Unanimous', 'R_win_by_KO/TKO', 'R_win_by_Submission',
       'R_win_by_TKO_Doctor_Stoppage', 'R_wins',
       'R_Height_cms',
       'R_Reach_cms', 'R_Weight_lbs', 'R_age', 'B_age', 'label']]
    
    X = df.drop(['label'], axis = 1).values
    y = df['label'].values
    
    return(X, y)

X_train, y_train = prep_ufc(train)
X_test,  y_test  = prep_ufc(test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import tree
fig, ax = plt.subplots(1)

best_score = 0

for d in range(1, 30):
    T = tree.DecisionTreeClassifier(max_depth = d)
    cv_score = cross_val_score(T, X_train, y_train, cv = 10).mean()

    ax.scatter(d, cv_score, color = 'black')

    if cv_score > best_score:
        best_depth = d
        best_score = cv_score

l = ax.set(title = "Best Depth : " + str(best_depth), xlabel = "Depth", ylabel = "CV Score")
```

```python
T = tree.DecisionTreeClassifier(max_depth = best_depth)
T.fit(X_train, y_train)
T.score(X_test, y_test)
```

```python
tree.plot_tree(T)
```

```python
cleaned = cleaned.dropna(axis = 1)
cleaned
from sklearn.model_selection import train_test_split

# np.random.seed(1234)
train, test = train_test_split(cleaned, test_size = 0.2) # hold out 20% of the data

train.shape, test.shape
```

```python
from sklearn import preprocessing
def preprocess(data_df):
    df = data_df.copy()
    df = df.drop(['R_odds', 'B_odds', 'R_ev', 'B_ev', 'date', 'location', 'country', 'isWeekend', 'days', 'month', 'year', 'class', 'favorite', 'bet_correct'], axis = 1)
    le1 = preprocessing.LabelEncoder()
    df['title_bout'] = le1.fit_transform(df['title_bout'])
    df['weight_class'] = le1.fit_transform(df['weight_class'])
    df['gender'] = le1.fit_transform(df['gender'])
    df['B_Stance'] = le1.fit_transform(df['B_Stance'])
    df['R_Stance'] = le1.fit_transform(df['R_Stance'])
    fighter_names = df[['R_fighter', 'B_fighter']]
    df = df.drop(['R_fighter', 'B_fighter'], axis = 1)
    X = df.drop(['Winner'], axis = 1).values
    y = df['Winner'].values
    
    X_cols = df.drop(['Winner'], axis = 1).columns
    X_cols = X_cols.tolist()
    
    return (X, y, X_cols, fighter_names)

X_train, y_train, cols, fighter_names = preprocess(train)
X_fighter_names = fighter_names
X_test,  y_test, cols, fighter_names = preprocess(test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import tree
fig, ax = plt.subplots(1)

best_score = 0

for d in range(1, 30):
    T = tree.DecisionTreeClassifier(max_depth = d)
    cv_score = cross_val_score(T, X_train, y_train, cv = 10).mean()

    ax.scatter(d, cv_score, color = 'black')

    if cv_score > best_score:
        best_depth = d
        best_score = cv_score

l = ax.set(title = "Best Depth : " + str(best_depth) + "; Best CV Score : " + str(best_score), xlabel = "Depth", ylabel = "CV Score")
```

```python
T = tree.DecisionTreeClassifier(max_depth = best_depth)
T.fit(X_train, y_train)
T.score(X_test, y_test)
```

```python
tree.plot_tree(T, filled = True, rounded = True)
```
